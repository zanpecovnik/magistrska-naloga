```python
import os
import json
import pandas as pd

from os.path import exists
```

```python
def parse_kas_data():
  data = []
  all_dirs = os.listdir(f"/content/drive/MyDrive/Colab Notebooks/magistrska/data/documents-kas2-json/sl2en")

  for i,dir in enumerate(all_dirs):
    print(f"Done with loading {i}/{len(all_dirs)} directories")

    all_files_in_dir = os.listdir(f"/content/drive/MyDrive/Colab Notebooks/magistrska/data/documents-kas2-json/sl2en/{dir}")

    for j, f in enumerate(all_files_in_dir):
      json_file = open(f"/content/drive/MyDrive/Colab Notebooks/magistrska/data/documents-kas2-json/sl2en/{dir}/{f}", "r", encoding='utf-8')
      single_file_json_data = json.load(json_file)

      single_file_data = {"text": [], "abstract_eng": ""}
      for key, arr in single_file_json_data.items():
        if key == "abstract":
          single_file_data["abstract_eng"] = arr[0]
        else:
          single_file_data["text"].extend(arr)

      single_file_data["text"] = " ".join(single_file_data["text"])
      data.append(single_file_data)
      json_file.close()

  return pd.DataFrame(data)

def save_df_to_drive(df: pd.DataFrame):
  df.to_csv(f"/content/drive/MyDrive/Colab Notebooks/magistrska/data/dataframe-kas.csv", encoding="utf-8", index=False)
```

```python
df = parse_kas_data()
df = df.dropna()
df = df.reset_index(drop=True)
```

```python
if not exists(f"/content/drive/MyDrive/Colab Notebooks/magistrska/data/dataframe-kas.csv"):
  print("Dataframe does not exist in drive, saving it...")
  save_df_to_drive(df)
```
